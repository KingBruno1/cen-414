```python
for year in range(2010, 2020):
    fig,ax = plt.subplots(1)
    data[str(2012)].plot.barh(figsize=(10,10), title=str(year))
```

```python
vals = []
for year in range(2010, 2020):
    vals.append(data[str(year)].mean())
    
plt.plot(range(2010,2020), vals)
plt.title('Mean')
```

```python
vals = []
for year in range(2010, 2020):
    vals.append(data[str(year)].median())  
    
plt.plot(range(2010,2020), vals)
plt.title('Median')
```

```python
vals = []
for year in range(2010, 2020):
    vals.append(data[str(year)].var())  
    
plt.title('Variance')
plt.plot(range(2010,2020), vals)
```

```python
vals = []
for year in range(2010, 2020):
    vals.append(data[str(year)].std())  
    

plt.title('Stabdard Deviation')    
plt.plot(range(2010,2020), vals)
```

```python
def iqr(sample):  
    return sample.quantile(0.75) - sample.quantile(0.25)


def siqr(sample):  
    return sample.quantile(0.75) - sample.quantile(0.25)


vals = []
for year in range(2010, 2020):
    vals.append(iqr(data[str(year)]))  
    
plt.title('Inter Quartile Range')
plt.plot(range(2010,2020), vals)

```

```python
vals = []
for year in range(2010, 2020):
    vals.append(siqr(data[str(year)]))  
    
plt.title('Semi Inter Quartile Range')
plt.plot(range(2010,2020), vals)
```
